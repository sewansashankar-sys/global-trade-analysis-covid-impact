```python
import pandas as pd

# Load data
df = pd.read_excel("Trade_Data.xlsx")

# 1. Drop completely empty columns
df = df.dropna(axis=1, how='all')

# 2. Drop irrelevant columns (you can adjust this)
cols_to_drop = [
    'typeCode', 'freqCode', 'refPeriodId',
    'reporterCode', 'partnerCode', 'partner2Code',
    'partner2ISO', 'partner2Desc',
    'classificationCode', 'classificationSearchCode',
    'isOriginalClassification', 'isLeaf',
    'customsCode', 'customsDesc',
    'mosCode', 'motCode',
    'qtyUnitCode', 'altQtyUnitCode',
    'isQtyEstimated', 'isAltQtyEstimated',
    'isNetWgtEstimated', 'isGrossWgtEstimated',
    'legacyEstimationFlag', 'isReported', 'isAggregate'
]

df = df.drop(columns=cols_to_drop, errors='ignore')

# 4. Rename columns for clarity
df = df.rename(columns={
    'reporterDesc': 'Country',
    'partnerDesc': 'partner_name',
    'cmdCode': 'product_code',
    'cmdDesc': 'product_description',
    'flowDesc': 'trade_flow',
    'motDesc': 'transport_mode',
    'primaryValue': 'trade_value',
    'refYear':'Year',
    'qty':'Quantity',
    'netWgt':'Net Weight'
})

# 5. Remove duplicates
df = df.drop_duplicates()

# 6. Convert types if needed
df['Year'] = df['Year'].astype(int)

# Save cleaned data
df.to_excel("Cleaned_TradeData.xlsx", index=False)

print("Data cleaned successfully!")
print(df.head())
#cleaning data changing features name into more detalied name
```

```python
# Remove extra spaces in column names
df.columns = df.columns.str.strip()

# Check missing values
print(df.isnull().sum())

# Drop or fill missing values
df = df.dropna()
#checking how many missing values are there in each feature
```

```python
df_pivot = df.pivot_table(
    values='Trade Value',
    index=['Country', 'Year'],
    columns='trade_flow'
).reset_index()

print(df_pivot.head())
#Checking how much india has imported or exported within the years
```

```python
df_pivot['Total_Trade'] = df_pivot['Export'] + df_pivot['Import']
#adding a new column for total trade which is the sum of import and export
```

```python
df_final = df.merge(df_pivot, on=['Country', 'Year'], how='left')

print(df_final.head())
#merging the new columns and sorting them according to country and year to decrease ambiguity
```
